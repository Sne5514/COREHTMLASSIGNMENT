```
<!DOCTYPE html>
<html lang="en">

<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Solution 4</title>
    <!-- 4. Create an ordered list of HTML tags. Each list item must include the tag name and some information about the tag. -->
</head>

<body>
    <h1>List of HTML Tags:</h1>
    <div>
        <ol>
            <ol><li><strong>&lt;html&gt;</strong>-  The root element that wraps all the content on the webpage. It is the container for everything in an HTML document.</li>
                <li><strong>&lt;head&gt;</strong> - Contains meta-information about the HTML document, like the title, character set, and links to external resources like stylesheets and scripts.</li>
                <li><strong>&lt;title&gt;</strong> - Defines the title of the document, which is displayed in the browser tab or title bar.</li>
                <li><strong>&lt;body&gt;</strong> - Contains the content of the HTML document, such as text, images, links, and other elements that are visible to the user.</li>
                <li><strong>&lt;h1&gt; to &lt;h6&gt;</strong> - Header tags used to define headings, with <strong>&lt;h1&
              gt;</strong> being the highest and most important level and <strong>&lt;h6&gt;</strong> being the lowest.</li>
                <li><strong>&lt;p&gt;</strong> - Represents a paragraph of text. It’s used to group related sentences together.</li>
                <li><strong>&lt;a&gt;</strong> - Defines a hyperlink. It allows you to link to other web pages or resources. The link is specified using the href attribute (e.g., <strong>&lt;a href="url"&gt;link text&lt;/a&gt;</strong>).</li>
                <li><strong>&lt;img&gt;</strong> - Embeds an image in the page. It uses attributes like <strong>src</strong> (image source) and <strong>alt</strong> (alternative text).</li>
                <li><strong>&lt;ul&gt;</strong> - Defines an unordered list, typically used for listing items with bullet points.</li>
                <li><strong>&lt;ol&gt;</strong> - Defines an ordered (numbered) list. Items inside it are usually numbered automatically.</li>
                <li><strong>&lt;li&gt;</strong> - Represents a list item. It can be used inside both ordered and unordered lists.</li>
                <li><strong>&lt;div&gt;</strong> - A container element used to group content together. It has no semantic meaning but is useful for styling and layout purposes.</li>
                <li><strong>&lt;span&gt;</strong> - A generic inline container used to style parts of text or group inline elements together without affecting their layout.</li>
                <li><strong>&lt;br&gt;</strong> - Inserts a line break, allowing text to continue on the next line.</li>
                <li><strong>&lt;strong&gt;</strong> - Defines important text, typically rendered in bold. It adds semantic meaning, indicating that the text is of strong importance.</li>
                <li><strong>&lt;em&gt;</strong> - Represents emphasized text, typically rendered in italics, and adds semantic meaning to emphasize content.</li>
                <li><strong>&lt;form&gt;</strong> - Defines an HTML form used for collecting user input. It can contain various input elements such as text fields, buttons, and checkboxes.</li>
                <li><strong>&lt;input&gt;</strong> - Defines an input field for a form, where users can enter data. It can have various types (e.g., text, password, checkbox).</li>
                <li><strong>&lt;button&gt;</strong> - Defines a clickable button, often used in forms or to trigger JavaScript actions.</li>
                <li><strong>&lt;table&gt;</strong> - Defines a table, which is used to display data in rows and columns.</li>
                <li><strong>&lt;tr&gt;</strong> - Represents a row in a table, containing one or more <strong>&lt;td&gt;</strong> (table data) elements.</li>
                <li><strong>&lt;td&gt;</strong> - Defines a cell in a table, where data is placed. It is used inside a <strong>&lt;tr&gt;</strong>.</li>
                <li><strong>&lt;th&gt;</strong> - Defines a header cell in a table. It is typically rendered in bold and centered to differentiate it from regular table data cells.</li>
                <li><strong>&lt;link&gt;</strong> - Defines a relationship between the HTML document and an external resource, often used to link to stylesheets.</li>
        </ol>
    </div>
</body>

</html>
```

```
<!DOCTYPE html>
<html lang="en">

<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Solution 5</title>
    <!-- 5. Create a description list of full stack web development tech stack, using the <dl> tag. Each term should be a tech stack name and each description should be a brief explanation of what the tech stack is used for. -->

</head>

<body>
    <h1>List of Full Stack Web Development tech stack:</h1>
    <dl>
      <dt>Frontend (Client-Side)</dt>
        <dd>Technologies used to build the part of the website that users interact with. It includes HTML, CSS, and JavaScript frameworks like React, Angular, and Vue.js.</dd>
       <dt>Backend (Server-Side)</dt>
        <dd>Technologies responsible for the server-side logic and database interaction. Common backend languages include Node.js, Python (Django, Flask), Ruby on Rails, and Java.</dd>
        <dt>Database</dt>
        <dd>Databases store and manage data for web applications. Common databases include SQL-based ones like MySQL, PostgreSQL, and NoSQL options like MongoDB.</dd>
        <dt>Version Control (Git)</dt>
        <dd>Version control systems like Git help developers track changes to code over time and collaborate with teams. GitHub, GitLab, and Bitbucket are popular platforms for code hosting and collaboration.</dd>
        <dt>Web Servers</dt>
        <dd>Web servers handle HTTP requests from clients (browsers) and send responses, typically using Apache, Nginx, or Node.js-based servers like Express.js.</dd>
        <dt>API (Application Programming Interface)</dt>
        <dd>APIs allow different software applications to communicate. RESTful APIs and GraphQL are common choices for building APIs that interact with the backend.</dd>
        <dt>Cloud Platforms</dt>
        <dd>Cloud platforms like AWS, Google Cloud, and Microsoft Azure are used to host web applications, databases, and server resources. They offer scalability and flexibility for deploying apps.</dd>
        <dt>DevOps Tools</dt>
        <dd>DevOps tools like Docker and Kubernetes are used for automating deployments, managing containerized applications, and ensuring continuous integration and delivery (CI/CD) pipelines.</dd>
        <dt>Testing</dt>
        <dd>Testing tools and frameworks such as Jest, Mocha, and Selenium are used to ensure code quality and verify that the application behaves as expected.</dd>
        <dt>Frameworks</dt>
        <dd>Full-stack frameworks like MERN (MongoDB, Express.js, React, Node.js) and MEAN (MongoDB, Express.js, Angular, Node.js) provide a structured approach to building full-stack applications.</dd>
    </dl>

</body>

</html>
```
